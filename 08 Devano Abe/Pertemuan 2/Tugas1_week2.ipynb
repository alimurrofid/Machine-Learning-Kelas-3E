```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

df = pd.read_csv('wbc.csv')
df.head()
```

1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan

```python
# mengetahui jumlah data yang hilang untuk setiap kolom
df.isnull().sum()
```

Output terlihat bahwa kolom "Unnamed: 32" memiliki 569 data yang hilang, sedangkan kolom lainnya tidak memiliki data yang hilang.

```python
# Menghapus kolom id
df = df.drop(['id', 'Unnamed: 32'], axis=1)
df.head()
```

## Penjelasan :

Pada kolom id dihapus karena tidak relevan untuk mendiagnosis kanker dan pada kolom Unnamed: 32 dihapus karena tidak berisi informasi apapun.

2. Lakukan proses encoding pada kolom "diagnosis"

```python
le = LabelEncoder() # membuat objek dari LabelEncoder
df['diagnosis'] = le.fit_transform(df['diagnosis']) # proses encoding
```

## Penjelasan :
Dilakukan proses encoding pada kolom "diagnosis" untuk mengubah nilai pada kolom "diagnosis" menjadi bilangan bulat. 
* membuat objek LabelEncoder dengan menggunakan kode **le = LabelEncoder()**
* Kategori  'M' (malignant) dan 'B' (benign) pada kolom "diagnosis" menjadi 'M' = 1 dan 'B' = 0 dengan kode **le.fit_transform** 

```python
df.head(10)
```

3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
std = StandardScaler()
numeric = df.columns.difference(['diagnosis'])
df[numeric] = std.fit_transform(df[numeric])
df.head(10)
```

## Penjelasan :
Dilakukan proses standarisasi pada semua kolom kecuali kolom diagnosis. 
* membuat objek StandardScaler dengan kode **std = StandardScaler()**.
* mendapatkan kolom numerik kecuali kolom "diagnosis" dengan kode **df.columns.difference** kemudian di simpan di variable numeric
* melakukan standardisasi pada kolom-kolom numerik dalam DataFrame yang diidentifikasi dan di simpan di variable numeric dengan kode **std.fit_transform** 
* tampilkan hasil yang telah distandarisasi sebanyak 10 teratas

4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
# Split data
from sklearn.model_selection import train_test_split

# Split data training dan dan lainnya
# data lainnya, akan kita split lagi menjadi validasi dan testing.
# Rasio yang akan kita gunakan adalah 8:2
df_train, df_unseen = train_test_split(df, test_size=0.2, random_state=0, stratify=df['diagnosis'])

# Cek rasio tiap label
print('===============================\n')
print(f'Jumlah label data asli\t\t:\n{df.diagnosis.value_counts()}')
print(f'\nJumlah label data latih/train\t:\n{df_train.diagnosis.value_counts()}')
print(f'\nJumlah label data uji/test\t:\n{df_unseen.diagnosis.value_counts()}')

# Cek masing-masing ukuran data
print('\n===============================')
print(f'\nJumlah data latih/train\t: {df_train.shape[0]}')
print(f'Jumlah data uji/test\t: {df_unseen.shape[0]}')
```

## Penjelasan :
Dilakukan proses split data untuk membuat data latih dan data uji dengan rasio 80:20
* Import modul **train_test_split** dari **sklearn.model_selection**.
* Menggunakan **train_test_split** untuk membagi DataFrame df menjadi data latih (df_train) dan data lainnya/uji (df_unseen). 
* Parameter **test_size** diatur sebagai 0.2, yang berarti 20% dari data akan digunakan sebagai data uji. 
* Parameter **random_state** diatur sebagai 0 untuk mengatur angka acak. 
* Parameter **stratify** diatur sebagai df['diagnosis'] untuk memastikan proporsi kelas yang sama di antara data latih dan data uji berdasarkan kolom 'diagnosis'.
* Kode **print(f'Jumlah label data latih/train:\n{df_train.diagnosis.value_counts()}')** digunakan untuk untuk mencetak jumlah label dalam data pelatihan setelah pembagian.
* Kode **print(f'Jumlah label data uji/test:\n{df_test.diagnosis.value_counts()}')** digunakan untuk untuk mencetak jumlah label dalam data pengujian setelah pembagian.
* Kode **print(f'Jumlah data latih/train: {df_train.shape[0]}')** digunakan untuk mencetak jumlah data dalam set pelatihan.
